```python
import requests
import pandas as pd
import time 
```

```python
def get_threads(key, thread, default='NaN'):
    return thread.get(key, default)

def get_all_threads():
    r = requests.get('https://a.4cdn.org/pol/catalog.json')
    r = r.json()
    
    data = []
    
    for page in r:
        threads = page['threads']
        
        for thread in threads:
            no = get_threads('no', thread)
            com = get_threads('com', thread)
            time = get_threads('time', thread)
            
            data.append({
                'id': no,
                'comment': com,
                'date': pd.to_datetime(time, unit='s')
            })
    
    return pd.DataFrame(data)
```

# Extract over a number of comments

```python
def get_4chan_comments(limit):
    run_loop = True

    while run_loop:
        df = pd.read_csv('../data/1 Politics/4chan/pol_4chan_comments.csv', sep=';')
        
        df_new = pd.concat([df, get_all_threads()])
        df_new = df_new.drop_duplicates()
        
        df_new.to_csv('../data/1 Politics/4chan/pol_4chan_comments.csv', index=False, sep=';')
        
        if len(df_new) >= limit:
            run_loop = False
        
        time.sleep(30)
```

```python
get_4chan_comments(10000)
```
