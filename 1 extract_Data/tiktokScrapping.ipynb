```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_comments(url):
    # Crear una instancia del controlador de Selenium con las opciones configuradas
    driver = webdriver.Chrome()

    driver.get(url)
    time.sleep(15)  # Esperar unos segundos para cargar la pÃ¡gina completamente

    # Hacer scroll hasta el final de la pÃ¡gina para cargar mÃ¡s tiktokComments
    prev_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")

        if new_height == prev_height:
            break

        prev_height = new_height

    # Obtener todo el HTML de la pÃ¡gina
    html = driver.page_source
    driver.quit()

    # Utilizar BeautifulSoup para analizar el HTML
    soup = BeautifulSoup(html, 'html.parser')

    tiktokComments = soup.find_all('p', class_="tiktok-1ht48ju-PCommentText")
    tiktokDates = soup.find_all('span', class_="tiktok-1esugaz-SpanCreatedTime e1g2efjf8")

    comments_data = {
        'comments': [],
        'date': []
    }

    for comment, date in zip(tiktokComments, tiktokDates):
        comments_data['comments'].append(comment.text)
        comments_data['date'].append(date.text)

    return pd.DataFrame(comments_data)
```

```python
url = 'https://www.tiktok.com/@samsung/video/7150486972560182529'

comments = get_comments(url)

comments.to_csv('../Data/tiktok/samsung_zflip_tk_comments.csv', sep=';', index=False)
comments
```
