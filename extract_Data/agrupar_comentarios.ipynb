```python
import pandas as pd
```

```python
def date(df, date_column):
    df[date_column] = pd.to_datetime(df[date_column])
    df[date_column] = df[date_column].dt.date
    return df
```

```python
def convert_unix_to_normal(df, unix_fecha_columna):
    df[unix_fecha_columna] = pd.to_datetime(df[unix_fecha_columna], unit='s').dt.strftime('%Y-%m-%d')
    return df
```

```python
def source(df, valor):
    df['source_id'] = valor
    return df
```

```python
def community(df, valor):
    df['community_id'] = valor
    return df
```

```python
df = pd.read_csv('../Data/4chan/pol_4chan_comments.csv', sep=';')

df = date(df, 'date')
df = source(df, 5)
df = community(df, "pol")

df.rename(columns={'id': 'publication_id', 'date' : 'date_data'}, inplace=True)



df = df[['source_id', 'community_id', 'publication_id', 'date_data', 'hate', 'comment']]

df
```
