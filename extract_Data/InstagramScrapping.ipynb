```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import pandas as pd
import re
```

### Cuenta

```python
def _get_dataframe(soup):
    instagramComments = soup.find_all('div', class_="_a9zo")
    comments_data = {
        'comment': [],
        'user': [],
        'date': []
    }
    
    for c in instagramComments:
        match = re.search(r'dir="auto">([^<]+)<', str(c))
        if match:
            comment = match.group(1)
            comments_data['comment'].append(comment)
        else:
            comments_data['comment'].append(None)

        filtered_elements = c(lambda tag: tag.name != 'svg' and tag.get_text(strip=True) != '')
        user = filtered_elements[2].text
        comments_data['user'].append(user)

        match = re.search(r'datetime="([^"]+)"', str(c))
        if match:
            date = match.group(1)
            comments_data['date'].append(date)
        else:
            comments_data['date'].append(None)

    return pd.DataFrame(comments_data)



def get_comments(url):
    # Crear una instancia del controlador de Selenium con las opciones configuradas
    driver = webdriver.Chrome()

    driver.get(url)
    time.sleep(60)  # Esperar unos segundos para cargar la página completamente

    html = driver.page_source
    driver.quit()

    soup = BeautifulSoup(html, 'html.parser')

    return _get_dataframe(soup)
```

```python
url = 'https://www.instagram.com/p/CugmYpDJ1b2/'

df = get_comments(url)
df.to_csv('../Data/instagram/samsung_last_ig_post_comments.csv', sep=";", index=False)

df
```
